```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

house=pd.read_csv("C:/Users/1/data2/practice/boston_housing.csv")
house.head()
```

```python
from scipy import stats
from scipy.stats import norm
#histogram and normal probability plot
sns.distplot(house['MEDV'].dropna(), fit=norm);
fig = plt.figure()
res = stats.probplot(house['MEDV'].dropna(), plot=plt)
```

```python
ax = sns.boxplot(y=house['MEDV'].dropna())
```

```python
#skewness and kurtosis
print("Skewness: " + str(house['MEDV'].dropna().skew()))
print("Kurtosis: " + str(house['MEDV'].dropna().kurt()))
```

```python
#applying log transformation
house['MEDV'] = np.log(house['MEDV'])
#transformed histogram and normal probability plot
sns.distplot(house['MEDV'].dropna(), fit=norm);
fig = plt.figure()
res = stats.probplot(house['MEDV'].dropna(), plot=plt)
```

```python
#skewness and kurtosis
print("Skewness: " + str(house['MEDV'].dropna().skew()))
print("Kurtosis: " + str(house['MEDV'].dropna().kurt()))
```
